```python

import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("/content/sample_data/diabetes.csv")
```

```python
dataset
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
scalar.fit(dataset.drop('Outcome',axis=1))
```

```python
scaled_features=scalar.transform(dataset.drop('Outcome',axis=1))
scaled_features
```

```python

new_dataset=pd.DataFrame(scaled_features,columns=dataset.columns[:-1])
new_dataset
```

```python
from sklearn.model_selection import train_test_split
X=new_dataset
Y=dataset['Outcome']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=95)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=80)
knn.fit(X_train,Y_train)
predict=knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(confusion_matrix(Y_test,predict))
```

```python
print(classification_report(Y_test,predict))
```

```python
print(accuracy_score(Y_test,predict))
```
